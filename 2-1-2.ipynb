```python
import requests
url = "http://data4library.kr/api/loanItemSrch?format=json&startDt=2022-04-01&endDt=2022-04-30&age=20&authKey=16f3c27a8ace4f54f403a9ff1177e6e5f234956a5f42a224b78fc7fff1eae1ba"
```

```python
r = requests.get(url)
data =  r.json()
print(data)
```

```python
data
```

```python
books = []
for d in data['response']['docs']:
    books.append(d['doc'])
```

```python
books
```

```python
import pandas as pd
books_df = pd.DataFrame(books)
books_df
```
